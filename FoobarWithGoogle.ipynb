```python
def solution(x):
    # Your code here
    decodedX = ""
    for i in x:
        if i.islower():
            newletter = (ord("z")-ord(i))+ord("a")
            decodedX += chr(newletter)
        else:
            decodedX += i
    return decodedX

print(solution("wrw blf hvv ozhg mrtsg'h vkrhlwv?"))
```
